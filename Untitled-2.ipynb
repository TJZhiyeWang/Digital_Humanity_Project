```python
import requests
import os
row = 0
col = 95
root = "../level20_1931/"
prefix = "https://wmts100.geo.admin.ch/1.0.0/ch.swisstopo.zeitreihen/default/19311231/2056/20/"
while True:
    os.mkdir(root + str(col))
    url = prefix + str(col) + "/" + str(row) +".png"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36"}
    html = requests.get(url, headers = headers, allow_redirects=False)
    if html.status_code == 400:
        break
    while True:
        url = prefix + str(col) + "/" + str(row) +".png"
        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.99 Safari/537.36"}
        html = requests.get(url, headers = headers, allow_redirects=False)

        if html.status_code == 200:
            file = root + str(col) + "/" + str(row) + ".png"
            with open(file,'wb')as f:
                f.write(html.content)
            print(col, row)
            row += 1
            continue
        else:
            col += 1
            row = 0
            break
```
